# Bubble Sort





```python
# the highest number will go to end automatically
# the second highgest number goes to last second row
nums=[-2, 3,11,45, 0, 11, -9]
for i in range(len(nums)): # 0,1,2,3,4  Decides the second logic use
    for j in range(len(nums)-i-1): # 0-4,0-3,0-2 Main logic to avoid last index number every time
        if nums[j]>nums[j+1]:
            nums[j],nums[j+1] = nums[j+1],nums[j] # Avoiding the use of temp by this line
    print(nums)

  
```

  # Selection Sort 
  
  

```python
# this will move minimum number to 1st index
# second minimum number to 2nd index
nums=[6, 4, 0, 1, 3]
for i in range(len(nums)): # 0 - 4
    minimum = i
    for j in range(i + 1, len(nums)): # 1-4 / 2-4 / 3-4 // leaving first index
        if nums[j] < nums[minimum]: # 4<6
            minimum = j
    nums[minimum], nums[i] = nums[i], nums[minimum] # # 
    print(nums)
```

# Quick sort



```python
# Quick sort in Python
data = [6, 7, 2, 1, 0, 9, 8]
size = len(data)

quickSort(data, 0, size - 1) # which is nothing but array , start and end index


# function to perform quicksort
def quickSort(array, low, high): # [8, 7, 2, 1, 0, 9, 6] , 0 , 6
  if low < high:

    # find pivot element such that ,element smaller than pivot are on the left, element greater than pivot are on the right
    pi = partition(array, low, high)

    # recursive call on the left of pivot
    quickSort(array, low, pi - 1)

    # recursive call on the right of pivot
    quickSort(array, pi + 1, high)
    
    
# function to find the partition position
def partition(array, low, high): # [8, 7, 2, 1, 0, 9, 6] , 0 , 6

  # choose the rightmost element as pivot
  pivot = array[high] # 6

  # pointer for greater element
  i = low - 1 # -1 this will point the last element

  # traverse through all elements compare each element with pivot
  for j in range(low, high): # 0-6
    if array[j] <= pivot:
      # if element smaller than pivot is found swap it with the greater element pointed by i
      i = i + 1

      # swapping element at i with element at j
      (array[i], array[j]) = (array[j], array[i])

  # swap the pivot element with the greater element specified by i
  (array[i + 1], array[high]) = (array[high], array[i + 1])

  # return the position from where partition is done
  return i + 1


print('Sorted Array in Ascending Order:')
print(data)

```
