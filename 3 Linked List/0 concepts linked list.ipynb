```python
# create custom linked list in python

class Node:
    def __init__(self,data=None):
        self.data=data
        self.next =None
        
class Linked:
    def __init__(self):
        self.head=None
        
    def show(self):
        node =self.head
        
        while node is not None:
            print(node.data)
            node=node.next 

    def add(self,  new):
        new_node=Node(new)
        new_node.next=self.head
        self.head =new_node
        
```

```python
link = Linked() # object to Linked class 
```

```python
elem = Node('Edureka') # add first element
link.head=elem
link.show()
```

```python
elem2=Node("python")  # add  element to begining of the list
link.head.next =elem2
link.show()
```
